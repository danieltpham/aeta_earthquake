```python
import pandas as pd
import numpy as np
import folium
```

```python
# Load
metadata = pd.read_csv('meta.csv')
target = pd.read_csv('target.csv')
```

```python
target['Time'] = pd.to_datetime(target['Time'])
print(target.shape)
target.head()
```

```python
print(metadata.shape)
metadata.head()
```

# Map time

```python
# Set up map
m = folium.Map(tiles="cartodbpositron")
m.fit_bounds([[22, 98], [34, 107]])
# Draw boundary of prediction area
folium.Rectangle([[22, 98], [34, 107]]).add_to(m)
```

```python
# Setup color for each year
col_dict = {'2016': 'blue', '2017': 'orange', '2018': 'black', '2019': 'green', '2020': 'purple'}

# Add past earthquake data (blue)
for i in range(190):
    coords = [target['Latitude'][i], target['Longitude'][i]]
    folium.CircleMarker(coords, 
                        popup = target['Time'][i],
                        radius = target['Magnitude'][i],
                        stroke = False,
                        fill = True,
                        fill_color=col_dict[str(target['Time'].dt.year[i])],
                        fill_opacity=0.3).add_to(m)
# Add station (red)    
for i in range(metadata.shape[0]):
    coords = [metadata['Latitude'][i], metadata['Longitude'][i]]
    folium.CircleMarker(coords, 
                        popup = metadata['StationID'][i],
                        radius = 0.5,
                        color='red').add_to(m)
```
